```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
data = {
    'Outlook': ['Sunny', 'Overcast', 'Rain', 'Sunny', 'Rain'],
    'Temperature': ['Hot', 'Hot', 'Mild', 'Cool', 'Mild'],
    'Humidity': ['High', 'High', 'High', 'Normal', 'Normal'],
    'Wind': ['Weak', 'Weak', 'Weak', 'Strong', 'Weak'],
    'Play': ['No', 'Yes', 'Yes', 'No', 'Yes']
}

df = pd.DataFrame(data)
print("Original Weather Dataset:")
print(df)
```

```python
# Fill missing values (if present)
df.fillna(df.mode().iloc[0], inplace=True)
```

```python
label_encoder = LabelEncoder()

for column in df.columns:
    df[column] = label_encoder.fit_transform(df[column])

print("\nEncoded Dataset:")
print(df)
```

```python
scaler = StandardScaler()

# Scale only feature columns (excluding target)
X = df.drop('Play', axis=1)
y = df['Play']

X_scaled = scaler.fit_transform(X)

print("\nScaled Feature Values:")
print(X_scaled)
```
